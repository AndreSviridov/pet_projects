Дана матрица m x n.

Нужно убедиться, что числа которая входят в одну диагональ одинаковые.

Написать функцию, которая принимает матрицу и возвращает True / False.

Пример 1.

Input: matrix = [[42, 7, 13, 99], [6, 42, 7, 13], [1, 6, 42, 7]]

Output: True

Пояснение: В матрице диагонали выглядят так:

[1]

[6, 6]

[42, 42, 42]

[7, 7, 7]

[13, 13]

[99]

Все числа в каждой диагонали одинаковые. Значит возвращаем True

Пример 2 Input: matrix = [[8, 23], [69, 1]]

Output: False

# Решение

```python
matrix = [[42, 7, 13, 99], [6, 42, 7, 13], [1, 6, 42, 7]]

list_all = []
for lists in matrix:
  for x in lists:
    list_all.append(x)
print(list_all)

diagonal_lists = []
for x in [x for x in range(len(list_all)) if x != len(matrix[0])-1]:
  current_list = []
  current_index = x
  while current_index < len(list_all):
    current_list.append(list_all[current_index])
    current_index += len(matrix[0]) + 1
  diagonal_lists.append(current_list)
print(diagonal_lists)

set_list = []
for x in diagonal_lists:
  set_list.append(set(x))
print(set_list)

final_result = True
for x in set_list:
  if len(x) != 1:
    final_result = False
    break
print(final_result)
```
